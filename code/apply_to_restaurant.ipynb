```python
import pickle
```

```python
with open('classifier_DTC.pkl') as f:
    classifier_DTC = pickle.load(f)
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('review_cleaned_restaurants.csv')
```

```python
df.columns[range(676, 678)]
```

```python
outcomes = ['IsFoodGood', 'IsServiceGood', 'IsAmbianceGood', 'IsDealsGood', 'IsPriceGood']
```

```python
extra_info = ['business_id', 'date']
```

```python
df.columns[range(676, 678)+range(668, 673)]
```

```python
X = df[df.columns.delete(range(676, 678)+range(668, 673))]
```

```python
result_DTC = classifier_DTC.predict(X)
```

```python
final_df_DTC = df.loc[:, extra_info]
```

```python
for i in range(len(outcomes)):
    final_df_DTC[outcomes[i]] = result_DTC[:, i]
```

```python
year = []
for i in range(len(final_df_DTC)):
    year.append(final_df_DTC.date[i].split('-')[0])
```

```python
final_df_DTC['date'] = year
```

```python
final_df_DTC
```

```python
final_df_DTC = final_df_DTC[final_df_DTC.date != '1']
```

```python
rests = list(set(final_df_DTC['business_id']))
```

```python
years = list(set(final_df_DTC['date']))
```

```python
rest = final_df_DTC[final_df_DTC['business_id'] == rests[2]]
```

```python
rest_history = {}
for year in years:
    year_data = rest[rest['date'] == year]
    review_num = float(len(year_data))
    if review_num == 0:
        continue
    rest_history[year] = {}
    for outcome in outcomes:
        rest_history[year][outcome] = sum(year_data[outcome]) / review_num
```

```python
rest_history
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rest_valid_years = sorted(rest_history.keys())
```

```python
for outcome in outcomes:
    plt.plot(rest_valid_years, [rest_history[year][outcome] for year in rest_valid_years], label=outcome)
plt.axis([int(rest_valid_years[0])-4, int(rest_valid_years[-1])+1, 0, 1])
plt.legend(loc='upper right', bbox_to_anchor=(1.4, 0.7))
plt.grid()
plt.xlabel('Year')
plt.ylabel('Percentage of positive review label')
plt.show()
```
